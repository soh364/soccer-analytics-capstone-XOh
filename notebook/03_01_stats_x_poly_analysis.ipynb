## Q1: Copa America Tournament Winner Market Efficiency

### Research Question
Do Polymarket tournament winner odds accurately reflect team quality as measured 
by aggregate xG performance?

### Key Findings

**1. Markets are highly efficient (r = 0.841, p < 0.001)**
- Strong positive correlation between xG difference and market odds
- xG explains 71% of variance in tournament winner pricing (R² = 0.707)
- Mean absolute mispricing: 4.9% (indicating limited arbitrage opportunities)

**2. No systematic biases detected**
- Defensive teams (low xG_against) not underpriced (p = 0.206)
- Markets do not overweight finishing luck (r = 0.170, p > 0.05)
- Suggests sophisticated market participants using advanced metrics

**3. Favorites correctly priced, contenders potentially undervalued**
- Argentina (winner): 65.5% odds vs 43.7% model prediction
  - Linear model underestimates at extreme xG values
  - Market correctly incorporated intangibles (Messi, experience, momentum)
- Mid-tier teams (Mexico, USA): 18.0% and 13.6% underpriced respectively
  - Potential value opportunities, but need validation against results

**4. Model limitations**
- Linear regression adequate for mid-range xG (-2 to +6)
- Breaks down at extremes (best/worst teams)
- Future work: Logistic regression for better handling of boundaries

### Conclusion
Polymarket efficiently prices Copa America tournament winners based on underlying 
xG performance. Limited arbitrage opportunities exist (~5 teams with >5% edge), 
primarily among mid-tier contenders that markets may systematically undervalue 
relative to favorites.

## Q2: Player Props Market Efficiency

### Research Question
Are Euro 2024 player prop odds (top scorer, player of tournament) efficiently 
priced relative to underlying xG performance?

### Key Findings

**1. Markets do not correlate with xG (R² = 0.005, p = 0.76)**
- Pre-tournament odds show NO relationship to player xG accumulation
- Contrast to Q1 (tournament winners: R² = 0.707)
- Mean absolute "mispricing": 1.9% (but mispricing concept is invalid here)

**2. Why xG fails to predict player prop odds**
- **Playing time uncertainty dominates**: Team advancement determines matches played
  - Early elimination: 3 matches (270 min)
  - Tournament winner: 7 matches (630 min)
  - 2.3x difference in xG accumulation opportunity
- **Pre-tournament unpredictability**: Cannot know which teams advance
- **xG/90 ignores this uncertainty**: High xG/90 meaningless if team exits early

**3. Market behavior suggests sophisticated understanding**
- Markets correctly de-emphasize individual xG
- Pricing reflects tournament structure complexity
- Low R² is EXPECTED, not inefficiency
- Random-looking odds reflect genuine uncertainty

**4. Comparison: Favorites vs Longshots**
- In post/live odds subset: Severe favorite bias (-35% overpricing)
- In pre-tournament subset: No systematic bias (insufficient data)
- Suggests live markets have behavioral biases, pre-tournament more efficient

### Conclusion
Player prop markets appear inefficient (R² = 0.005) but this reflects inherent 
unpredictability, not market failure. Unlike tournament winners (deterministic 
xG aggregation), player props depend on unknowable team advancement paths. 
Markets correctly price this uncertainty by largely ignoring individual xG 
in favor of team success probabilities.

**Implication**: Player top scorer bets are closer to lottery tickets than 
value investing opportunities. Pre-tournament xG models offer minimal edge.

## Q3: Copa America Match-Level Exploratory Analysis

### Research Question (Exploratory)
In 6 Copa America knockout matches with available markets, how well did 
pre-match odds correlate with actual xG generated?

### Methodology
- Sample: 6 Copa America 2024 knockout matches
- Compared: Pre-match Polymarket odds vs actual match xG
- Metrics: Prediction accuracy (did favorite win?), correlation

### Key Findings

**1. xG outperformed market odds (4/6 vs 2/6 correct)**
- xG correctly predicted winner in 67% of cases
- Market odds correctly predicted winner in 33% of cases
- However, n=6 too small for statistical significance

**2. Market behavior: Default to 50/50 in knockouts**
- 4/6 markets priced at exactly 50/50 odds
- Even when xG showed clear favorite (Argentina 3.54 vs Canada 0.98)
- Suggests markets either: (a) genuinely uncertain, or (b) lazy pricing

**3. Clear mismatches priced correctly**
- Colombia vs Panama: 100% odds, 5-0 result, 2.18-0.61 xG
- Market efficient on obvious quality gaps

**4. Subtle edges missed**
- Argentina-Canada: 50% odds despite massive xG advantage (3.54 vs 0.98)
- Potential value betting opportunity

**5. Weak correlation (r = 0.345, p = 0.503)**
- No statistical relationship between odds and xG
- Either markets using non-xG signals or sample too small

### Case Studies

**Argentina vs Canada (Semifinals)**
- Market: 50/50 | xG: 3.54-0.98 | Result: 2-0 Argentina
- Analysis: Market mispriced Canada's "Cinderella run" narrative
- xG correctly identified quality gap

**Argentina vs Colombia (Final)**
- Market: 50/50 | xG: 1.08-0.99 | Result: 1-0 Argentina
- Analysis: Market correctly priced as toss-up, xG confirmed

### Limitations
- **Sample size: n=6 (critically underpowered)**
- Cannot draw statistical conclusions
- Findings are exploratory observations only
- Correlation not significant (p=0.503)
- Need 20+ matches for meaningful inference

### Conclusion
Exploratory findings suggest Copa America knockout markets may default 
to 50/50 pricing, potentially missing xG-based edges. However, sample 
too small to confirm. Future work should expand to larger match dataset 
for robust analysis.
